#### 2. What is the annual attrition rate of the company overall and when you slice the data on different dimensions? Does any particular segment stand out (why) ?

```python
# Convert 'Attrition_Date' to datetime if it's not already in datetime format
df['Attrition_Date'] = pd.to_datetime(df['Attrition_Date'], errors='coerce')

# Extract year from 'Attrition_Date'
df['Attrition_Year'] = df['Attrition_Date'].dt.year

# Calculate the total number of employees at the beginning of each year
employee_count_per_year = df.groupby(df['Attrition_Year'])['EmployeeID'].nunique()

# Calculate the number of employees who left the company within a year
attrition_count_per_year = df.groupby(df['Attrition_Year'])['EmployeeID'].count()

# Calculate the annual attrition rate overall
annual_attrition_rate_overall = attrition_count_per_year.sum() / employee_count_per_year.min()

# Calculate the annual attrition rate for each year
annual_attrition_rate_per_year = attrition_count_per_year / employee_count_per_year

# Slice the data on different dimensions (e.g., Department, Gender, JobRole) and calculate the annual attrition rate
annual_attrition_rate_by_dimension = df.groupby(['Attrition_Year', 'Department'])['EmployeeID'].count() / df.groupby(['Attrition_Year', 'Department'])['EmployeeID'].nunique()

# Repeat the above step for other dimensions as needed

# Print results
print("Annual Attrition Rate Overall:", annual_attrition_rate_overall)
print("\nAnnual Attrition Rate by Year:")
print(annual_attrition_rate_per_year)
print("\nAnnual Attrition Rate by Dimension:")
print(annual_attrition_rate_by_dimension)
```

#### 3. What all factors impact attrition the most (statistical analysis expected) ? Are there any early signs which can help us be more proactive ? (E.g. Satisfaction dropping, Last rating not good)

```python
# Convert 'Attrition_Date' to numeric representation (number of days since a reference date)
df['Attrition_Date_numeric'] = (df['Attrition_Date'] - df['Attrition_Date'].min()).dt.days

# List of numerical columns
num_cols = ['DistanceFromHome', 'MonthlyIncome', 'NumCompaniesWorked', 'PercentSalaryHike', 'YearsAtCompany', 
            'YearsSinceLastPromotion', 'YearsWithCurrManager', 'BreaksTakenPerDay', 'EnvironmentSatisfaction', 
            'JobSatisfaction', 'WorkLifeBalance', 'JobInvolvement', 'PerformanceRating']

# Calculate the correlation coefficients
correlation_results = df[num_cols].corrwith(df['Attrition_Date_numeric'])

# Print correlation coefficients
print("Correlation with Attrition (numeric representation of date):")
print(correlation_results)
```

#### 7. The HR Head has a opinion that people who have changed a lot of jobs in past are the ones with very high attrition and should be the biggest focus of retention program. How will you test this hypothesis ?

```python
from scipy.stats import chi2_contingency

# Group the data by 'NumCompaniesWorked' and calculate attrition rate for each group
attrition_by_companies = df.groupby('NumCompaniesWorked')['Attrition_Date'].mean()

# Print attrition rates for each group
print("Attrition Rate by Number of Companies Worked For:")
print(attrition_by_companies)

# Hypothesis testing (chi-square test)
observed = pd.crosstab(df['Attrition_Date'], df['NumCompaniesWorked'])
chi2, p, dof, expected = chi2_contingency(observed)

# Print chi-square test results
print("\nChi-square test results:")
print("Chi-square statistic:", chi2)
print("p-value:", p)
```

#### The hypothesis is valid as it indicates that people who have changed jobs a lot are higher prone to attrition.

#### 1. Calculate the average time duration spent by each employee using In Time & Out Time

```python
import warnings

warnings.filterwarnings("ignore")

merged_time = pd.merge(in_time, out_time, on='EmployeeID')

# Calculate the time duration spent by each employee for each day
for date in in_time.columns[1:]:  # Exclude the first column (EmployeeID)
    in_time_col = str(date) + '_x'  # Convert date to string
    out_time_col = str(date) + '_y'  # Convert date to string
    merged_time[date] = pd.to_datetime(merged_time[out_time_col]) - pd.to_datetime(merged_time[in_time_col])
    merged_time[date] = merged_time[date].dt.total_seconds() / 3600  # Convert to hours

# Calculate the average time duration spent by each employee across all days
merged_time['AverageTime(hrs)'] = merged_time.iloc[:, 1:].mean(axis=1)

# Display the resulting dataframe with average time duration
print(merged_time[['EmployeeID', 'AverageTime(hrs)']])
```

#### 2. Create Bins based on the average time duration into below bins
#### a. High: If employee spent more than 8 Hours
#### b. Medium: if employee spent more than 6 Hours
#### c. Low: If employee spent less than 6 Hours

```python
# Define the bins and labels
bins = [0, 6, 8, 20]  # Adjusted bin edges
labels = ['Low', 'Medium', 'High']

# Create a new column 'TimeCategory' based on the average time duration bins
merged_time['TimeCategory'] = pd.cut(merged_time['AverageTime(hrs)'], bins=bins, labels=labels, right=False)

# Display the resulting dataframe with time category
print(merged_time[['EmployeeID', 'AverageTime(hrs)', 'TimeCategory']])
```
